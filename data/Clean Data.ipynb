```python
import pandas as pd
import numpy as np
```

# Air Pollution Data

```python
air_quality_2019 = pd.read_csv('daily_aqi_by_cbsa_2019.csv')
air_quality_2020 = pd.read_csv('daily_aqi_by_cbsa_2020.csv')
air_quality_2021 = pd.read_csv('daily_aqi_by_cbsa_2021.csv')
```

```python
frames = [air_quality_2019,  air_quality_2020, air_quality_2021]
air_quality = pd.concat(frames)
air_quality
```

```python
air_quality_by_date = air_quality.groupby('Date').mean()['AQI']
air_quality_by_date.head()
```

```python
air_quality_by_date.to_csv('air_quality_by_date.csv')
```

# Trips by Distance Data

```python
trips_by_distance = pd.read_csv('Trips_by_Distance.csv')
```

```python
trips_by_distance['Level'].head()
```

```python
trips_by_distance_national = trips_by_distance.loc[trips_by_distance['Level'] == 'National']
trips_by_distance_national = trips_by_distance_national.reset_index()
trips_by_distance_national = trips_by_distance_national.drop(['index', 'Level', 'State FIPS', 'State Postal Code', 'County FIPS', 'County Name', 'Row ID', 'Week', 'Month'],  axis=1)
trips_by_distance_national
```

```python
trips_by_distance_national.to_csv('trips_by_distance_national.csv', index=False)
```
