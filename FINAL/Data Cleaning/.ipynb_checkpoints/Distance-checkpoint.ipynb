```python
import numpy as np
import pandas as pd
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('distance2.csv')
df = df.set_index('Date')
```

```python
df = df.reset_index()
df
```

```python
x_filtered = df[["Number of Trips <1"]].apply(savgol_filter,  window_length=51, polyorder=2)
x = df[["Number of Trips <1"]]
plt.ion()
plt.plot(x)
plt.plot(x_filtered)
plt.show()
```

```python
df['Number of Trips 1-5'] = df['Number of Trips 1-3'] + df['Number of Trips 3-5']
df['Number of Trips 5-25'] = df['Number of Trips 5-10'] + df['Number of Trips 10-25']
df['Number of Trips 25-100'] = df['Number of Trips 50-100'] + df['Number of Trips 25-50']
df['Number of Trips >=250'] = df['Number of Trips >=500'] + df['Number of Trips 250-500']
df = df[['Date','Number of Trips <1', 'Number of Trips 1-5', 'Number of Trips 5-25', 'Number of Trips 25-100','Number of Trips 100-250','Number of Trips >=250']]
df
```

```python
df['Date'] = pd.to_datetime(df['Date']) + pd.to_timedelta(7, unit='d')
df1 = df.groupby(pd.Grouper(freq='W', key='Date'))['Number of Trips <1', 'Number of Trips 1-5', 'Number of Trips 5-25', 'Number of Trips 25-100','Number of Trips 100-250','Number of Trips >=250'].sum()
```

```python
compression_opts = dict(method='zip',
                        archive_name='distance_out.csv')  
df.to_csv('out.zip', index=False,
          compression=compression_opts) 
```
