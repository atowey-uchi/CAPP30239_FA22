```python
df = pd.read_csv('mobility.csv', na_values='National', dtype={'County FIPS': 'string'})
```

```python
df = df[df['County FIPS'] != '<NA>']
df
```

```python
df['Percent at Home'] = df['Population Staying at Home']/(df['Population Staying at Home']+df['Population Not Staying at Home'])
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df[['County FIPS', 'Date', 'Percent at Home']]
```

```python
df_march19 = df[df['Date']=='2019/03']
df_feb20 = df[df['Date']=='2020/02']
df_march20 = df[df['Date']=='2020/03']
df_april20 = df[df['Date']=='2020/04']
df_may20 = df[df['Date']=='2020/05']
df_march21 = df[df['Date']=='2021/03']
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv')  
df.to_csv('out.zip', index=False,
          compression=compression_opts)  
```

```python
import zipfile
from zipfile import ZipFile

with zipfile.ZipFile("outfiles.zip", "w") as zf:
      with zf.open(f"march19.csv", "w") as buffer:
         df_march19.to_csv(buffer, index=False)
      with zf.open(f"feb20.csv", "w") as buffer:
         df_feb20.to_csv(buffer, index=False)
      with zf.open(f"march20.csv", "w") as buffer:
         df_march20.to_csv(buffer, index=False)
      with zf.open(f"april20.csv", "w") as buffer:
         df_april20.to_csv(buffer, index=False)
      with zf.open(f"may20.csv", "w") as buffer:
         df_may20.to_csv(buffer, index=False)
      with zf.open(f"march21.csv", "w") as buffer:
         df_march21.to_csv(buffer, index=False)
```
